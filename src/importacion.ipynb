* Módulo con el kmeans

```python
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, adjusted_rand_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder, MinMaxScaler


#Metodo del Codo y Silueta para encontrar el numero optimo de clusters sin alojamientos turísticos
df_num_sin=df_num.iloc[:, :10]

X=df_num_sin

X=StandardScaler().fit_transform(X)

wcss = []
ss = []
for i in range(2, 11):
    #Metodo del Codo para encontrar el numero optimo de clusters
    kmeans = KMeans(n_clusters = i, random_state = 42)
    kmeans_labels = kmeans.fit_predict(X)
    wcss.append(kmeans.inertia_)
   #Metodo de la silueta para encontrar el numero optimo de clusters
    silhouette_avg = silhouette_score(X, kmeans_labels)
    ss.append(silhouette_avg)

#Grafica método del codo
plt.figure(figsize=(15,10))
plt.plot(range(2, 11), wcss)
plt.title('Método del codo')
plt.xlabel('Número de clusters')
plt.ylabel('WCSS')
plt.show()

#Grafica método del silueta
plt.figure(figsize=(15,10))
plt.plot(range(2, 11), ss)
plt.title('Método de la silueta')
plt.xlabel("Número de clusters")
plt.ylabel("Silhouette Score")
plt.show()


#Calculamos kmeans para k = 4
kmeans_sin = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans_sin = kmeans_sin.fit_predict(X)


print("\nSilhouette_score kmeans sin alojamientos turísticos: ", silhouette_score(X, y_kmeans_sin))

#Añadimos cluster al dataset
df_sin = df.iloc[:, :12]
df_sin["Cluster_sin_pipeline"] = y_kmeans_sin + 1

#kmeans con Pipeline
preprocessor = Pipeline(
    [
        ("scaler", MinMaxScaler()),
        ("pca", PCA(n_components=2, random_state=42)),
    ]
)

clusterer = Pipeline(
    [
        (
            "kmeans",
            KMeans(
                n_clusters=4,
                init="k-means++",
                n_init=50,
                max_iter=500,
                random_state=42,
            ),
        ),
    ]
)

pipe = Pipeline(
    [
        ("preprocessor", preprocessor),
        ("clusterer", clusterer)
    ]
)

silhouette_scores = []
for n in range(2, 11):
    # This set the number of components for pca,
    # but leaves other steps unchanged
    pipe["preprocessor"]["pca"].n_components = n
    pipe.fit(df_num_sin)

    silhouette_coef = silhouette_score(
        pipe["preprocessor"].transform(df_num_sin),
        pipe["clusterer"]["kmeans"].labels_ + 1,
    )

    # Add metrics to list
    silhouette_scores.append(silhouette_coef)

plt.style.use("fivethirtyeight")
plt.figure(figsize=(6, 6))
plt.plot(range(2, 11),silhouette_scores,c="#008fd5",label="Coeficiente Silhouette")

plt.xlabel("n_components")
plt.legend()
plt.title("Rendimiento del clustering en función de los n_components sin alojamientos turísticos")
plt.tight_layout()
plt.show()

pipe["preprocessor"]["pca"].n_components = 2
pipe.fit(df_num_sin)

preprocessed_data = pipe["preprocessor"].transform(df_num_sin)
predicted_labels = pipe["clusterer"]["kmeans"].labels_ + 1

print("Silhouette_score pipeline sin alojamientos turísticos: ", silhouette_score(preprocessed_data, predicted_labels))

pcadf = pd.DataFrame(
    pipe["preprocessor"].transform(df_num_sin),
    columns=["component_1", "component_2"],
)

pcadf["predicted_cluster"] = pipe["clusterer"]["kmeans"].labels_ + 1
plt.style.use("fivethirtyeight")
plt.figure(figsize=(8, 8))

scat = sns.scatterplot(
    "component_1",
    "component_2",
    s=50,
    data=pcadf,
    hue="predicted_cluster",
    palette="Set2",
)

scat.set_title("Resultado Kmeans sin alojamientos turísticos")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
plt.show()

componentes=pd.DataFrame(
        data    = pipe["preprocessor"]["pca"].components_,
        columns = df_num_sin.columns,
        index   = ['PC1', 'PC2']
    )

print('---------------------------------------------------------------------------------------')
print('Peso de las variables en cada componente del PCA con alojamientos turísticos (PC1, PC2)')
print('---------------------------------------------------------------------------------------')
display(componentes)



# Heatmap componentes
# ==============================================================================
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(4, 4))
plt.imshow(componentes.T, cmap='viridis', aspect='auto')
plt.yticks(range(len(df_num_sin.columns)), df_num_sin.columns)
plt.xticks(range(len(pipe["preprocessor"]["pca"].components_)), np.arange(pipe["preprocessor"]["pca"].n_components_) + 1)
plt.grid(False)
plt.colorbar();
plt.title("Mapa de calor de las variables en cada componente del PCA sin alojamientos turísticos")
plt.show()

modelo_pca=pipe["preprocessor"]["pca"]

# Porcentaje de varianza explicada por cada componente
# ==============================================================================
print('--------------------------------------------------------------------------------')
print('Porcentaje de varianza explicada por cada componente sin alojamientos turísticos')
print('--------------------------------------------------------------------------------')
print(modelo_pca.explained_variance_ratio_)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))
ax.bar(
    x      = np.arange(modelo_pca.n_components_) + 1,
    height = modelo_pca.explained_variance_ratio_
)

for x, y in zip(np.arange(len(df_num_sin.columns)) + 1, modelo_pca.explained_variance_ratio_):
    label = round(y, 2)
    ax.annotate(
        label,
        (x,y),
        textcoords="offset points",
        xytext=(0,10),
        ha='center'
    )

ax.set_xticks(np.arange(modelo_pca.n_components_) + 1)
ax.set_ylim(0, 1.1)
ax.set_title('Porcentaje de varianza explicada por cada componente sin alojamientos turísticos')
ax.set_xlabel('Componente principal')
ax.set_ylabel('Por. varianza explicada');
plt.show()

df_sin["Cluster"] = pipe["clusterer"]["kmeans"].labels_ + 1

df_cluster_sin = df_sin.sort_values(['Distrito','Cluster'])

#Mostramos dataset con el cluster
display('Mostramos dataset con el cluster sin alojamiento turístico')
display(df_cluster_sin)

#Metodo del Codo y Silueta para encontrar el numero optimo de clusters teniendo en cuenta los alojamientos turísticos
X=df_num

X=StandardScaler().fit_transform(X)

wcss = []
ss = []
for i in range(2, 11):
    # Metodo del Codo para encontrar el numero optimo de clusters
    kmeans = KMeans(n_clusters = i, random_state = 42)
    kmeans_labels = kmeans.fit_predict(X)
    wcss.append(kmeans.inertia_)
   # Metodo de la silueta para encontrar el numero optimo de clusters
    silhouette_avg = silhouette_score(X, kmeans_labels)
    ss.append(silhouette_avg)

# Grafica método del codo
plt.figure(figsize=(15,10))
plt.plot(range(2, 11), wcss)
plt.title('Método del codo')
plt.xlabel('Número de clusters')
plt.ylabel('WCSS')
plt.show()
# Grafica método del silueta
plt.figure(figsize=(15,10))
plt.plot(range(2, 11), ss)
plt.title('Método de la silueta')
plt.xlabel("Número de clusters")
plt.ylabel("Silhouette Score")
plt.show()


#Calculamos kmeans para k = 4
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)

print("Silhouette_score kmeans con alojamientos turísticos: ", silhouette_score(X, y_kmeans))

#Añadimos cluster al dataset
df["Cluster_sin_pipeline"] = y_kmeans + 1

#kmeans con Pipeline
preprocessor = Pipeline(
    [
        ("scaler", MinMaxScaler()),
        ("pca", PCA(n_components=2, random_state=42)),
    ]
)

clusterer = Pipeline(
    [
        (
            "kmeans",
            KMeans(
                n_clusters=4,
                init="k-means++",
                n_init=50,
                max_iter=500,
                random_state=42,
            ),
        ),
    ]
)

pipe = Pipeline(
    [
        ("preprocessor", preprocessor),
        ("clusterer", clusterer)
    ]
)

silhouette_scores = []
for n in range(2, 11):
    # This set the number of components for pca,
    # but leaves other steps unchanged
    pipe["preprocessor"]["pca"].n_components = n
    pipe.fit(df_num)

    silhouette_coef = silhouette_score(
        pipe["preprocessor"].transform(df_num),
        pipe["clusterer"]["kmeans"].labels_ + 1,
    )

    # Add metrics to list
    silhouette_scores.append(silhouette_coef)

plt.style.use("fivethirtyeight")
plt.figure(figsize=(6, 6))
plt.plot(range(2, 11),silhouette_scores,c="#008fd5",label="Coeficiente Silhouette")

plt.xlabel("n_components")
plt.legend()
plt.title("Rendimiento del clustering en función de los n_components con alojamientos turísticos")
plt.tight_layout()
plt.show()

pipe["preprocessor"]["pca"].n_components = 3
pipe.fit(df_num)

preprocessed_data = pipe["preprocessor"].transform(df_num)
predicted_labels = pipe["clusterer"]["kmeans"].labels_ + 1

print("Silhouette_score pipeline con alojamientos turísticos PCA 3 componentes: ", silhouette_score(preprocessed_data, predicted_labels))


pipe["preprocessor"]["pca"].n_components = 2
pipe.fit(df_num)

preprocessed_data = pipe["preprocessor"].transform(df_num)
predicted_labels = pipe["clusterer"]["kmeans"].labels_ + 1

print("Silhouette_score pipeline con alojamientos turísticos PCA 2 componentes: ", silhouette_score(preprocessed_data, predicted_labels))

pcadf = pd.DataFrame(
    pipe["preprocessor"].transform(df_num),
    columns=["component_1", "component_2"],
)

pcadf["predicted_cluster"] = pipe["clusterer"]["kmeans"].labels_ + 1
plt.style.use("fivethirtyeight")
plt.figure(figsize=(8, 8))

scat = sns.scatterplot(
    "component_1",
    "component_2",
    s=50,
    data=pcadf,
    hue="predicted_cluster",
    palette="Set2",
)

scat.set_title("Resultados Kmeans con alojamientos turísticos")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
plt.show()

componentes=pd.DataFrame(
        data    = pipe["preprocessor"]["pca"].components_,
        columns = df_num.columns,
        index   = ['PC1', 'PC2']
    )

print('--------------------------------------------------------------------------------')
print('Peso de las variables en cada componente del PCA con alojamientos turísticos')
print('--------------------------------------------------------------------------------')
display(componentes)



# Heatmap componentes
# ==============================================================================
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(4, 4))
plt.imshow(componentes.T, cmap='viridis', aspect='auto')
plt.yticks(range(len(df_num.columns)), df_num.columns)
plt.xticks(range(len(pipe["preprocessor"]["pca"].components_)), np.arange(pipe["preprocessor"]["pca"].n_components_) + 1)
plt.grid(False)
plt.colorbar();
plt.title("Mapa de calor de las variables en cada componente del PCA con alojamientos turísticos")
plt.show()

modelo_pca=pipe["preprocessor"]["pca"]

# Porcentaje de varianza explicada por cada componente
# ==============================================================================
print('-------------------------------------------------------------------------------------------')
print('Porcentaje de varianza explicada por cada componente con alojamientos turísticos (PC1, PC2)')
print('-------------------------------------------------------------------------------------------')
print(modelo_pca.explained_variance_ratio_)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))
ax.bar(
    x      = np.arange(modelo_pca.n_components_) + 1,
    height = modelo_pca.explained_variance_ratio_
)

for x, y in zip(np.arange(len(df_num.columns)) + 1, modelo_pca.explained_variance_ratio_):
    label = round(y, 2)
    ax.annotate(
        label,
        (x,y),
        textcoords="offset points",
        xytext=(0,10),
        ha='center'
    )

ax.set_xticks(np.arange(modelo_pca.n_components_) + 1)
ax.set_ylim(0, 1.1)
ax.set_title('Porcentaje de varianza explicada por cada componente con alojamientos turísticos')
ax.set_xlabel('Componente principal')
ax.set_ylabel('Por. varianza explicada');
plt.show()

df["Cluster"] = pipe["clusterer"]["kmeans"].labels_ + 1

df_cluster = df_sin.sort_values(['Distrito','Cluster'])

#Mostramos dataset con el cluster
display('Mostramos dataset con el cluster con alojamiento turístico')
display(df_cluster)


```
